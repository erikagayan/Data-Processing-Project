```python
import pandas as pd
```

```python
df = pd.read_csv("db_table.csv", index_col=[0])
```

```python
df.head()
```

```python
def calculate_bitcode(source):
    return 1 if source == 'S1' else 0

df['bitCode'] = df['source'].apply(calculate_bitcode)
df
```

```python
def calculate_sicode(bit_code, code3):
    if bit_code == 1:
        if code3 in ['AP', 'AH']:
            return 'A'
        elif code3 == 'PRD':
            return 'B'
        elif code3 == 'YLD':
            return 'BpA'
    else:
        if code3 in ['AP', 'AH']:
            return 'H'
        elif code3 == 'PRD':
            return 'T'
        elif code3 == 'YLD':
            return 'TpH'
    return None

df['siCode'] = df.apply(lambda row: calculate_sicode(row['bitCode'], row['code3']), axis=1)
df
```

```python
def transform_G_code1(G_code1):
    return G_code1

def transform_G_code2(G_code2):
    return G_code2.split("/")[0] if "/" in G_code2 else G_code2
```

```python
unique_code1 = df["code1"].unique()
unique_code1
```

```python
unique_code2 = df["code2"].apply(lambda x: x.split("/")[0]).unique()
unique_code2
```

```python
from itertools import product

possible_combinations = list(product(unique_code1, unique_code2))
possible_combinations
```

```python
sources_dict = {}

for comb in possible_combinations:
    transformed_code1 = transform_G_code1(comb[0])
    transformed_code2 = transform_G_code2(comb[1])
    
    filtered_data = df[(df["code1"].str.startswith(transformed_code1)) & 
                         (df["code2"].str.startswith(transformed_code2))]
    
    unique_sources = filtered_data["source"].unique().tolist()
    if unique_sources:
        sources_dict[comb] = unique_sources

sources_dict
```

```python
grouped_data_dict = {}

for (G_code1, G_code2), sources in sources_dict.items():
    for source in sources:
        filtered_grouped_data = df[(df['code1'].str.startswith(transform_G_code1(G_code1))) & 
                                     (df['code2'].str.startswith(transform_G_code2(G_code2))) & 
                                     (df['source'] == source)]
        selected_data = filtered_grouped_data[['updateDate', 'code1', 'code2', 'code3', 'value', 'siCode', 'bitCode']]
        grouped_data_dict[(G_code1, G_code2, source)] = selected_data.to_dict('records')

list(grouped_data_dict.items())[:5]
```
